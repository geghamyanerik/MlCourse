```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x = np.array([3,5,4,6]).reshape((-1, 1))
y = np.array([8,16,15,20])
```

```python
plt.scatter(x,y)
plt.grid()
```

$\beta = (X^TX)^{-1}X^TY$

```python
def linear_regression(x, y, return_intercept=True):
    if len(x.shape) == 1:
        x = x.reshape((-1, 1))
    if return_intercept:
        x = np.concatenate([np.ones((x.shape[0], 1)), x], axis=1)
    y = y.reshape((-1, 1))
    beta = np.linalg.inv(x.T @ x) @ x.T @ y
    return beta
```

```python
beta = linear_regression(x, y)

beta
```

```python
x1 =np.linspace(0,10,10)
y1 = beta[0] + beta[1] * x1
plt.scatter(x,y)
plt.plot(x1,y1)
plt.grid()
```

```python
y_p = beta[0] + beta[1] * x
mse = np.mean((y-y_p)**2)
mse
```

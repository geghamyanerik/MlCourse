**14.** Какую долю (вещественное число от 0 до 1, округлить до 4-го знака) от всех покупателей составляют ВМЕСТЕ мужчины от 26 до 35 лет и женщины старше 36 лет (то есть нужно учесть несколько возрастных категорий)? (речь не об уникальных ID, а о количестве таких строк)  

```python
# Ваш код здесь
m = data[(data["Gender"] == "M")& (data["Age"].between("26","35"))]
f = data[(data["Gender"] == "F")& (data["Age"] >= "36")]
len_data = len(data)

res = (len(m) + len(f))/len_data
round(res,4)
```

```python
# ответ: 0.4077
```
